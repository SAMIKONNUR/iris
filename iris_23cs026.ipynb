```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='/Iris.csv'
df=pd.read_csv('/content/Iris.csv')
color={'Iris-setosa':'red','Iris-virginica':'blue' , 'Iris-versicolor':'green'}
sp_color = df['Species'].map(color)
df.columns=['Id', 'sepal_Length','sepal_Width','petal_Length','petal_Width','class']
print(df.columns)
x = df['petal_Length']
y = df['petal_Width']
plt.scatter(x,y, c=sp_color)
plt.xlabel('sepal_Length in CM')
plt.ylabel('sepal_Width in CM')
plt.title('scater plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name = '/content/Iris.csv'
df = pd.read_csv(file_name)
species_count = df['Species'].value_counts()
plt.bar(species_count.index, species_count.values, color='lightblue')
plt.title("Number of flowers by species")
plt.xlabel('Species')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
